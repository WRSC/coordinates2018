```python
import folium
from glob import glob
from IPython.display import HTML, display
import os.path as osp
import pandas
import yaml
```

```python
calshot_latlon = [50.820483, -1.311233]
```

```python
markers = pandas.read_csv('../area_scanning.csv', index_col=0)
markers
```

```python
start_icon = folium.Icon(color='green', icon='play')
```

```python
[m[1:] for m in markers.itertuples()]
```

```python
def plot_track(csv_file):
    boat_trace = pandas.read_csv(csv_file)
    
    # Create the map
    map_osm = folium.Map(location=calshot_latlon, zoom_start=17)
    
    # Add the corner markers
    for m in markers.itertuples():
        
        folium.Marker(m[1:], popup=m.Index).add_to(map_osm)

    # Add the boat track
    latlons = [(row['field.latitude'], row['field.longitude']) for (ix, row) in boat_trace.iterrows()
               if (row['field.latitude'] > 50.7) and (row['field.longitude'] > -2) and (row['field.longitude'] < -1.3)]
    # Select 10% of points to make the notebook a bit smaller
    latlons = latlons[::10]
    map_osm.add_child(folium.features.PolyLine(latlons))
    
    # Mark which end is the start 
    start = boat_trace.iloc[0]
    folium.Marker(latlons[0], icon=start_icon).add_to(map_osm)
    return map_osm
```

```python
plot_track('plymouth-own-gps/pb_first_attempt.csv')
```

```python
plot_track('plymouth-own-gps/ps_first_attempt.csv')
```
