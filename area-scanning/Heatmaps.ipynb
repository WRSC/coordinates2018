```python
import folium
from glob import glob
from IPython.display import HTML, display
import json
import os.path as osp
import pandas
import yaml
```

```python
calshot_latlon = [50.820483, -1.311233]
```

```python
markers = pandas.read_csv('../area_scanning.csv', index_col=0)
markers
```

```python
corners = [(markers.loc[p].lat, markers.loc[p].lon) for p in 'ABDC']
area_polygon = folium.features.PolygonMarker(corners, fill_opacity=0.)
```

```python
with open('heat_map_datamaster_micro_sailboat.csv.json') as f:
    micro_data = json.load(f)
max(s[1] for s in micro_data)
```

```python
# Create the map
map_osm = folium.Map(location=calshot_latlon, zoom_start=17)

# Add the corner markers
map_osm.add_child(area_polygon)

for corners, nboats in micro_data:
    folium.features.PolygonMarker(corners, fill_opacity=(0.3 * nboats)).add_to(map_osm)

map_osm
```
