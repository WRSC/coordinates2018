# Comparing GPS

Comparing two GPS traces for Southampton's boat

```python
import folium
import pandas
import yaml
```

```python
map_osm = folium.Map(location=[50.821116, -1.310633], zoom_start=17)
```

Load waypoints from YAML and add to map

```python
markers = pandas.read_csv('../fleet_race.csv', index_col=0)

for name, latlon in markers.iterrows():
    folium.Marker(latlon, popup=name).add_to(map_osm)
```

Load GPS from CSV and add to map

```python
competition_tracker = pandas.read_csv('Southampton2.csv')

latlons = [(row.latitude, row.longitude) for row in competition_tracker.itertuples()]
map_osm.add_child(folium.features.PolyLine(latlons));
```

Load the GPS from the boat's own tracking, compare it with the official tracking.

* Blue: official
* Green: team tracker

```python
team_tracker = pandas.read_csv('Southampton2.csv.ownrecord', names=['time', 'lat', 'long'])

latlons_selfrec = [(row.lat / 1e7, row.long / 1e7) for row in team_tracker.itertuples()]
map_osm.add_child(folium.features.PolyLine(latlons_selfrec, color='#0a2'))
```
